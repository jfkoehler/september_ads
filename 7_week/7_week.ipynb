```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

*Univariate Quantitative*:

- Histograms
- Boxplots
- Density Plots



```python
fig, ax = plt.subplots()
ax.hist(penguins['bill_length_mm'])
```

```python
fig, ax = plt.subplots(1, 2)
penguins['bill_length_mm'].plot(kind = 'hist', ax = ax[1])
```

```python
fig, ax = plt.subplots(nrows = 1, ncols=3, figsize = (15, 6))

ax[0].hist(penguins['bill_length_mm'], color = 'yellow');
penguins['bill_length_mm'].hist(ax = ax[1], color = 'red')
penguins['bill_length_mm'].plot(kind = 'hist', ax = ax[2], color = 'green')
```

```python
plt.boxplot(penguins['bill_length_mm'].dropna());
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15, 6))
ax[0, 0].hist(penguins['bill_depth_mm'])
ax[0, 0].set_title('Bill Depth')

ax[0, 1].hist(penguins['bill_length_mm'], color = 'salmon', edgecolor = 'black')
ax[0, 1].set_title('Bill Length')

ax[1, 0].boxplot(penguins['flipper_length_mm'].dropna())

ax[1, 1].hist(penguins['body_mass_g']);
```
