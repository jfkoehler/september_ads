```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
```

```python
#normal distributions
standard_normal = stats.norm()
```

```python
#generate 100 random variables
standard_normal.rvs(100)
```

```python
standard_normal.pdf(0)
```

```python
x = np.linspace(-4, 4, 1000)
```

```python
plt.plot(x, standard_normal.pdf(x))
```

```python
standard_normal.cdf(0)
```

```python
plt.plot(x, standard_normal.cdf(x))
```

```python
d1 = stats.norm(loc = 5, scale = 10)
d2 = stats.norm(loc = 7, scale = 9)
```

```python
x = np.linspace(-30, 40, 1000)
plt.plot(x, d1.pdf(x))
plt.plot(x, d2.pdf(x))
```

```python
stats.ttest_1samp(d1.rvs(1000), 2.8)
```

```python
stats.ttest_ind(d1.rvs(1000), d2.rvs(1000))
```
