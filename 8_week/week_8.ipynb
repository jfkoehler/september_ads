```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
X, y = cancer.data, cancer.target
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(X, columns=cancer.feature_names)
```

```python
df['target'] = y
```

```python
df.head()
```

```python
#start small
zeros = df.loc[df['target'] == 0]
ones = df.loc[df['target'] == 1]
```

```python
columns = df.columns[:-1]
```

```python
columns
```

```python
len(columns)
```

```python
for i in range(3):
    for j in range(10):
        print(i, j)
```

```python
fig, axs = plt.subplots(15,2, figsize = (18, 30))
col_counter = 0
for i in range(15):
    for j in range(2): 
        zeros[columns[col_counter]].hist(ax = axs[i, j], color = 'purple', edgecolor = 'black',alpha = 0.3)
        ones[columns[col_counter]].hist(ax = axs[i, j], color = 'grey', edgecolor = 'black', alpha = 0.3)
        axs[i, j].set_title(columns[col_counter])
        col_counter += 1
        
fig.suptitle('Features Broken by Cancerous or Not',fontsize = 16)
plt.tight_layout()
```

```python
from scipy import stats
```

```python
normal_dist.rvs(10)
```

```python
binomial_dist = stats.binom(n = 100, p = 0.5)
```

```python
binomial_dist.rvs(10)
```

```python
binomial_dist.cdf(50)
```

```python
binomial_dist.pmf(45)
```
